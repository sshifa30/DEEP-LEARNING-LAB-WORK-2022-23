Spliting of training and testing data

```python
# dataset is imbalanced so balanced it
# so create two dataframe with df_major and df_minor
df_major = dataset[(dataset['capital-gain'] == 0)]
df_minor = dataset[(dataset['capital-gain'] == 1)]
df_minority_upsampled = resample(df_minor,
                                 replace=True,    # sample with replacement
                                 n_samples= 10945, # to match majority class
                                 random_state= 7255)  # result to reproduce
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_minority_upsampled, df_major])
```

```python
x = df_upsampled.drop("capital-gain", axis = 1) # axis = 1 removes column and axis = 0 removes row from dataset
y = df_upsampled["capital-gain"]
x_train, x_test, y_train, y_test = train_test_split(x,y)
```
